```python
def salary_category(salary):
    if salary < 50000:
        return "Low"
    elif 50000 <= salary < 80000:
        return "Medium"
    else:
        return "High"
```

```python
df["Salary"].apply(salary_category)
```

```python
df["salary_category"] = df["Salary"].apply(salary_category)
```

```python
df
```

```python
def adjust_performance(experience):
    if experience > 10:
        return 1
    else:
        return 0
```

```python
df["Experience"].apply(adjust_performance)
```

```python
df["Adjusted"] = df["Experience"].apply(adjust_performance)
df
```

```python
df["New_Score"] = df["Performance_Score"] + df["Adjusted"]
df
```

```python
df.drop(["Adjusted", "New_Score"], axis = 1, inplace = True)
```

```python
def adjust_new(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    else:
        return row["Performance_Score"]
```

```python
df["Adjusted_Performance"] = df.apply(adjust_new, axis = 1)
df
```

```python
df["Formated_Name"] = df["Name"].apply(lambda x: x.replace("_", " "))
df
```

```python
df.drop(["Adjusted_Performance","Formated_Name"], axis = 1, inplace = True)
df
```
