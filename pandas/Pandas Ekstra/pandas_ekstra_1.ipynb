```python
# merge
```

```python
df_merge1 = pd.read_csv("7-merge_data1.csv")
df_merge1
```

```python
df_merge2 = pd.read_csv("7-merge_data2.csv")
df_merge2
```

```python
# how parametresi ile birlestirme turunu belirleyebiliriz (inner, left, right, outer)
# outer : tum degerleri alir, eksik olan yerlere NaN degeri atar
df_merged_outer = pd.merge(df_merge1, df_merge2, on = "Employee_ID", how = "outer")
df_merged_outer
```

```python
# inner : sadece ortak olan degerleri alir
df_merged_inner = pd.merge(df_merge1, df_merge2, on = "Employee_ID", how = "inner")
df_merged_inner
```

```python
# left : sol dataframe deki tum degerleri alir, sag dataframe deki eksik degerler NaN olur
df_merged_left = pd.merge(df_merge1, df_merge2, on = "Employee_ID", how = "left")
df_merged_left
```

```python
# right : sag dataframe deki tum degerleri alir, sol dataframe deki eksik degerler NaN olur
df_merged_right = pd.merge(df_merge1, df_merge2, on = "Employee_ID", how = "right")
df_merged_right
```
