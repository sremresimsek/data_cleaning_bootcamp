```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure() # figur olusturmak icin kullanilir
```

```python
plt.figure(figsize = (12,8)) # figur boyutunu ayarlamak icin kullanilir
```

```python
# plt.plot
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10, 12]

plt.figure(figsize = (4, 4)) # yeni bir figur olusturur ve boyutunu ayarlar
# figsize ne kadar kucuk olursa grafikteki detaylar o kadar az gorunur

plt.plot(x, y) # x ve y degerlerini kullanarak bir cizgi grafigi cizer
```

```python
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = arr1 * (-3)

plt.figure(figsize = (6, 4))
plt.plot(arr2, arr1)
```

```python
arr3 = np.array([1, 2, 3, 4, 5])
arr4 = arr3 ** 2

plt.figure(figsize = (6, 4))
plt.plot(arr4, arr3, color = "black", linestyle = "--") # renk ve cizgi stili ayarlanabilir
```

```python
arr5 = np.arange(6)
arr6 = 2 * arr5 + 3

plt.figure(figsize = (6, 4))
plt.plot(arr5, arr6, color = "purple", marker = ".", markersize = 10, markerfacecolor = "orange") 
# marker = nokta isaretleyici ekler, isaretleyici boyutu ve rengi ayarlanabilir
```

```python
print(arr5)
print(arr6)
```

```python
arr7 = np.arange(100) / 10
arr8 = np.sin(arr7)

plt.plot(arr7, arr8) # sin fonksiyonunun grafigini cizer
```

```python
arr9 = np.arange(6)
arr10 = arr9 * 2
arr11 = arr9 ** 2

plt.figure(figsize = (8, 6))
plt.plot(arr9, arr10, label = "arr9 * 2")
plt.plot(arr9, arr11, label = "arr9 ** 2")
# iki farkli veriyi ayni grafige cizer
plt.legend(loc = "center left") # grafige legend ekler, hangi cizginin neyi temsil ettigini gosterir, loc ile konumu ayarlanabilir
plt.show() # grafigi gosterir 
```

```python
# plt.scatter
```

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [0, 2, 4, 6, 8, 10, 12]

plt.figure(figsize = (4, 4))

plt.plot(x, y, color = "cyan") # x ve y degerlerini kullanarak bir cizgi grafigi cizer
plt.scatter(x, y, color = "magenta") # x ve y degerlerini kullanarak bir serpilme (scatter) grafigi cizer, noktalari isaretler
# plt.plot ile arasindaki fark, plt.plot cizgi grafigi cizerken, plt.scatter nokta grafigi cizer
```

```python
arr12 = np.arange(100) / 10
arr13 = np.sin(arr12)
arr14 = np.cos(arr12)

plt.plot(arr12, arr13, color = "green", label = "sin(x)")
plt.scatter(arr12, arr14, color = "red", label = "cos(x)")
plt.legend() # default olarak en iyi (best) konuma yerlestirir
```

## plt.plot ve plt.scatter parametreleri

### plt.plot
`color` : cizginin rengini belirler

`linestyle` : cizgi stilini belirler

`marker` : noktalara denk gelen kesisimleri gosterir

`markersize, markerfacecolor` : isaretcinin rengini ve boyutunu belirler

`alpha` : 1 olursa opak, 0 olursa saydam, arasinda bir deger alabilir

`label` : cizginin etiketini belirler

### plt.scatter
`color` : cizginin rengini belirler

`s` : noktalarin boyutunu belirler

`edgecolors` : noktalarin kenarlarina renk ekler

`linewidths` : noktalarin kenarlarindaki renklerin boyutunu belirler

`alpha` : 1 olursa opak, 0 olursa saydam, arasinda bir deger alabilir

`label` : cizginin etiketini belirler
