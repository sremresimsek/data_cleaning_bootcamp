```python
sns.set_style("whitegrid") # seaborn ile beyaz arka plan ve izgara cizgileri ekleniyor
sns.scatterplot(x = "Height", y = "Weight", data = df, hue = "Sex") # hue parametresi ile cinsiyet bilgisi ekleniyor
plt.xlabel("Yukseklik")
plt.ylabel("Kilo")
plt.title("Yukseklik ve Kilo Arasindaki Iliski")
```

```python
sns.set_style("whitegrid") # seaborn ile beyaz arka plan ve izgara cizgileri ekleniyor
sns.scatterplot(x = "Height", y = "Weight", data = df, hue = "Sex", style = "Medal", size = "Age") 
# hue parametresi ile cinsiyet bilgisi ekleniyor, style parametresi ile madalya bilgisi ekleniyor, size parametresi ile yas bilgisi ekleniyor
plt.xlabel("Yukseklik")
plt.ylabel("Kilo")
plt.title("Yukseklik ve Kilo Arasindaki Iliski")
```

```python
sns.set_style("whitegrid") # seaborn ile beyaz arka plan ve izgara cizgileri ekleniyor
sns.lineplot(x = "Height", y = "Weight", data = df, hue = "Sex")
plt.xlabel("Yukseklik")
plt.ylabel("Kilo")
plt.title("Yukseklik ve Kilo Arasindaki Iliski")
```

```python
sns.set_style("whitegrid") # seaborn ile koyu arka plan ve izgara cizgileri ekleniyor
sns.displot(x = "Height", hue = "Sex", data = df)
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
```

```python
sns.set_style("whitegrid") # seaborn ile koyu arka plan ve izgara cizgileri ekleniyor
sns.displot(x = "Height", hue = "Sex", data = df,kind = "kde")
plt.ylabel("Frequency")
plt.title("Athletes Height Distribution")
```

```python
sns.barplot(x = "Medal", y = "Height", hue = "Sex", data = df)
plt.title("Medals by Height")
```

```python
sns.catplot(x = "Medal", y = "Height", hue = "Sex", col = "Season", data = df)
```

```python
sns.catplot(x = "Medal", y = "Height", hue = "Sex", col = "Season", data = df, kind = "point")
```

```python
df_titanic = sns.load_dataset("titanic")
df_titanic
```

```python
plt.figure(figsize = (12, 7))

plt.subplot(1, 2, 1)
sns.boxplot(x = "class", y = "age", data = df_titanic)
plt.title("Age by Class")
plt.xlabel("Class")
plt.ylabel("Age")

plt.subplot(1, 2, 2)
sns.boxplot(x = "class", y = "fare", data = df_titanic)
plt.title("Fare by Class")
plt.xlabel("Class")
plt.ylabel("Fare")

plt.tight_layout()
```
