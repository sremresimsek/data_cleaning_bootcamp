```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
age_list = [10, 20, 30, 30, 30, 40, 50, 60, 70, 75]
weight_list = [20, 60, 80, 85, 86, 87, 77, 90, 80, 90]
```

```python
plt.plot(age_list, weight_list, "r") # "r" : kirmizi, "y" sari, "b" : mavi, "g" : yesil
plt.xlabel("Age") # x eksenini isimlendir
plt.ylabel("Weight") # y eksenini isimlendir
plt.title("Age vs Weight") # grafiğin başlığını belirle
plt.show() # grafigi goster
# plt.savefig("test.png") # png turunde kaydet
```

```python
arr1 = np.linspace(0, 10, 20)
arr1
```

```python
arr2 = arr1 ** 3
```

```python
arr2
```

```python
plt.plot(arr1, arr2, "g-*") 
# "b*" : mavi yıldız işareti ile çizim yapar 
# "b*-" : mavi yıldız işareti ile çizim yapar ve çizgileri birleştirir
# "b--" : mavi kesik çizgi ile çizim yapar
# "b-." : mavi noktalı kesik çizgi ile çizim yapar
```

```python
plt.subplot(1, 2, 1)
plt.plot(arr1, arr2, "r--")
plt.subplot(1, 2, 2)
plt.plot(arr2, arr1, "y--")
# bu komutlar ile tek bir pencerede iki farkli grafik cizilir
```

```python
my_figure = plt.figure() # yeni bir figure (sekil) olusturur
figure_axes = my_figure.add_axes([0.1, 0.1, 0.4, 0.4]) # figure icerisinde eksenler olusturur
# eksenlerin konumu ve boyutu [sol, alt, genislik, yukseklik] seklindedir (0 ile 1 arasinda degerler alir)
figure_axes.plot(arr1, arr2, "g")
figure_axes.set_xlabel("X Axis")
figure_axes.set_ylabel("Y Axis")
figure_axes.set_title("Graph in Figure")
```

```python
new_fig = plt.figure(dpi = 100) # dpi = dots per inch (inc basina nokta sayisi), dpi arttikca grafik kalitesi artar
new_axes = new_fig.add_axes([0.1, 0.1, 0.6, 0.6]) 
new_axes.plot(arr1, arr1 ** 2, label = "arr1 ** 2") # veri seti cizilir ve etiket eklenir
new_axes.plot(arr1, arr1 ** 3, label = "arr1 ** 3") # iki farkli veri seti cizilir
new_axes.legend(loc = 9) # legend ekler (etiketler icin) # loc parametresi ile legend'in konumu belirlenir (0-10 arasinda deger alir)
plt.show()
```

```python
my_fig, my_axes = plt.subplots()
my_axes.plot(data_1, data_2, "purple") # cizgi renklerini istege gore belirleyebiliriz
my_axes.plot(data_2, data_1, "black", alpha = 0.5) # alpha parametresi ile renk saydamligi ayarlanir (0 ile 1 arasinda deger alir)
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()

new_axes.plot(data_1, data_1, "#8800FF", lw = 3.0, ls = ":") # lw parametresi ile cizgi kalinligi ayarlanir
new_axes.plot(data_1, data_1 + 1, "#000000", lw = 1.0, ls = "--", marker = "o", markersize = 8, markerfacecolor = "red") # ls parametresi ile cizgi stili ayarlanir
new_axes.plot(data_1, data_1 + 2, "#000000", lw = 0.5, ls = "-.", marker = "+", markersize = 4) # marker parametresi ile isaret tipi belirlenir
```
