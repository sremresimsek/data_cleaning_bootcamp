```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_1 = {
    "Yil" : np.arange(2010, 2021),
    "Satislar" : np.random.randint(100, 500, 11),
    "Maliyetler" : np.random.randint(50, 400, 11),
}

df_1 = pd.DataFrame(data_1)
df_1
```

```python
df_1["Bolgeler"] = np.random.choice(["Kuzey", "Guney", "Dogu", "Bati"], df_1.shape[0])
# bolgeler = ["Kuzey", "Guney", "Dogu", "Bati"]
# df_1["Bolge"] = np.random.choice(bolgeler, df_1.shape[0])
df_1
```

```python
plt.plot(df_1["Yil"], df_1["Satislar"],
         label = "Yıllara Göre Satışlar",
         color = "purple",
         linestyle = "--",
         marker = "o"
)

plt.legend()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_1["Yil"], df_1["Maliyetler"], color = "red", marker = "x")
plt.title("Yıllara Göre Maliyetler")
plt.xlabel("Yıllar")
plt.ylabel("Maliyetler") 
```

```python
plt.figure(figsize=(10, 6))
plt.plot(df_1["Yil"], df_1["Maliyetler"], label = "Yıllara Göre Maliyetler", color = "black", marker = "x")

plt.plot(df_1["Yil"], df_1["Satislar"], label = "Yıllara Göre Satışlar", color = "orange", marker = "o")

plt.title("Yıllara Göre Satışlar ve Maliyetler")
plt.legend()
plt.show()
```

### plt.bar

`plt.bar()` : sutun grafigi

```python
kategori = ["A", "B", "C", "D", "E"]
degerler = [23, 45, 12, 36, 28]

plt.bar(kategori, degerler, color = "skyblue", width = 0.5, edgecolor = "blue", linewidth = 2) 
# width: sutun genisligi, edgecolor: kenar rengi, linewidth: kenar kalinligi
plt.title("Kategoriye Göre Değerler")
```

```python
data_2 = {
    "Şehir" : ["Ankara", "İstanbul", "İzmir", "Bursa", "Antalya"],
    "Nüfus" : [5503985, 15029231, 4367251, 3056120, 2444535],
}

df_2 = pd.DataFrame(data_2)

df_2
```

### plt.hist

```python
plt.hist(df_1["Satislar"], bins = 10, alpha = 0.5, color = "purple", label = "Satışların Frekansı")
plt.legend()
plt.show()
```

```python
# bins : histogramda kaç adet çubuk olacağını belirler (grup sayisi)
# range: histogramın kapsayacağı değer aralığı
# density : True -> Grafik yoğunluk belirtir, False -> Grafik frekans belirtir
data_3 = np.random.randn(1000)

plt.hist(data_3, bins = 30, color = "cyan", range = (-4, 4), density = False, alpha = 1, label = "Frekans")
plt.legend()
plt.show()
```

```python
plt.hist(data_3, bins = 30, color = "magenta", range = (-4, 4), density = True, alpha = 1, label = "Yoğunluk") 
plt.legend()
plt.show()
```
