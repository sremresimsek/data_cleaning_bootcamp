```python
for col in df.columns:
    print("Col:", col)
    mask = df[col].isnull()
    
    if mask.any():
        print("\tNulls detected in column:", col)
        mode_value = df[col].mode().iloc[0]
        print("\t\tFilling nulls with mode value:", mode_value)
        df[col].fillna(mode_value, inplace=True)
```

```python
df["date_added"] = pd.to_datetime(df["date_added"])
df["year_added"] = df["date_added"].dt.year
df["month_added"] = df["date_added"].dt.month
df.head(2)
```

```python
df[df["type"] == "Movie"]
```

```python
df[df["type"] == "Movie"].groupby(["year_added", "month_added"]).size().reset_index(name ="count")
```

```python
movie_counts = df[df["type"] == "Movie"].groupby(["year_added", "month_added"]).size().reset_index(name ="count")
```

```python
heatmap_data = movie_counts.pivot_table(values = "count", index = "year_added", columns = "month_added", fill_value = 0)
sns.heatmap(heatmap_data, cmap = "coolwarm")
plt.title("Movies Added to Netflix Over Time")
plt.show()
```

```python
release_count = df.groupby(["release_year", "type"]).size().reset_index(name = "count")

sns.lineplot(data = release_count, x = "release_year", y = "count", hue = "type")
plt.title("Content Released on Netflix Over Time")
plt.show()
```
